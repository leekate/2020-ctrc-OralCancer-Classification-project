# DataEnhancement-Canny Filter

```img_canny = cv2.Canny(image, threshold1, threshold2, edges=None, apertureSize=None, L2gradient=None)```
* 첫번째 아규먼트 image는 입력 이미지

* 두번째, 세번째 아규먼트 threshold1, threshold2는 최소 스레숄드와 최대 스레숄 

* 네번째 아규먼트 edges에 Canny 결과를 저장할 변수.  파이썬에선 Canny 함수 리턴으로 받을 수 있기 때문에 필요없다. 

* 다섯번째 아규먼트 apertureSize는 이미지 그레디언트를 구할때 사용하는 소벨 커널 크기. 디폴트는 3

* 여섯번째 아규먼트 L2gradient가 True이면 그레디언트 크기를 계산할 때 sqrt{(dI/dx)^2 + (dI/dy)^2}를 사용

 False라면 근사값인 |dI/dx|+|dI/dy|를 사용. 디폴트값은 False. 

````
import cv2
from google.colab.patches import cv2_imshow

#img_gray = cv2.imread('/content/precan_3.jpg_0_4734.jpg', cv2.IMREAD_GRAYSCALE)
img_gray = cv2.imread('/content/precan_3.jpg_0_4734.jpg')
img_gray = cv2.resize(img_gray, (300, 300))   
cv2_imshow(img_gray)

img_canny = cv2.Canny(img_gray, 5, 150)
cv2_imshow(img_canny)
````

````
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt


img_ori = cv2.imread('/content/precan_3.jpg_0_4734.jpg')
img_ori = cv2.cvtColor(img_ori,cv2.COLOR_BGR2RGB)
#img_ori = cv2.resize(img_ori, (300, 300))   
img = cv2.imread('/content/precan_3.jpg_0_4734.jpg',0)
equ=cv2.equalizeHist(img)
img_canny = cv2.Canny(equ, 30, 300)


image4 = cv2.imread('/content/precan_3.jpg_0_4734.jpg', cv2. IMREAD_GRAYSCALE)
for i in range (0, 220):
    for j in range (0, 220):
        if image4[i,j] > 155:
            image4[i,j] = 255
            continue
        image4[i,j] += 100



plt.figure(figsize=(20,4))
imgae1=plt.subplot(1,5,1)
imgae1.set_title('Original')
plt.axis('off')
plt.imshow(img_ori)

imgae3=plt.subplot(1,5,2)
imgae3.set_title('gray')
plt.axis('off')
plt.imshow(img, cmap="gray")

imgae2=plt.subplot(1,5,3)
imgae2.set_title('equalization')
plt.axis('off')
plt.imshow(equ,cmap="gray")

imgae3=plt.subplot(1,5,4)
imgae3.set_title('canny')
plt.axis('off')
plt.imshow(img_canny,cmap="gray")

imgae4=plt.subplot(1,5,5)
imgae4.set_title('image luminance control')
plt.axis('off')
plt.imshow(image4,cmap="gray")



plt.show()
````
